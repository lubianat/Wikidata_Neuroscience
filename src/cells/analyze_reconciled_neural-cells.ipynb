### Loading OpenRefine reconciled dataset

```python
import pandas as pd

recon_neuro = pd.read_csv("../../data/cells/reconciled_neuralcells.csv")
recon_neuro.head()
```

```python
recon_neuro['IRI'] = recon_neuro['IRI'].str.replace("obo.", "")
recon_neuro.head()
```

Assembling quickstatements for adding instance of cell type and Cell Ontology ID

```python
for index, row in recon_neuro.iterrows():
        print(
      row['qid']+ "|P31|" + "Q189118" + '\n' +
      row['qid'] + "|P7963|" + '"'+ row['IRI'] + '"')
```
