Getting a list of neural disorders to create new items/improve existing items in Wikidata

```python
import requests
import re
from bs4 import BeautifulSoup as BS
```

Scraping the full list of disorders from [NINDS](https://www.ninds.nih.gov/), which is the same reference Wikipedia uses for [their list](https://en.wikipedia.org/wiki/List_of_neurological_conditions_and_disorders#References)

```python
links = []
for p in range(23):
    url = f'https://www.ninds.nih.gov/Disorders/All-Disorders?title=&page={p}'
    response = requests.get(url)
    soup = BS(response.content, "html.parser")

    links.extend(soup.find_all('a', {'href': re.compile(r'/Disorders/All-Disorders/.*')}))
```

```python
disorders = [link.get_text().replace(' Information Page', '') for link in links]
disorders
```

Saving resulting list to a csv for reconciling with OpenRefine

```python
import pandas as pd
disorder_list = pd.DataFrame({'Disorder': disorders})
disorder_list.to_csv("../../data/disorders/neuro-disorders_list.csv")
```
