In this notebook I acquire the disease ontology IDs for the neural disorders in the reconciled list.

```python
from owlready2 import *
import pandas as pd
```

```python
do = get_ontology("https://raw.githubusercontent.com/DiseaseOntology/HumanDiseaseOntology/master/src/ontology/HumanDO.owl").load()
```

```python
neuro = pd.read_csv('../../data/disorders/recon-neuro-disorders_list.csv')
disord = neuro.copy(deep=True)
```

```python
doid = []
for lab in disord['Disorder']:
    doid.append(do.search_one(label = lab))
disord['doid'] = [str(j) for j in doid]
disord
```

```python
ind = disord['doid'].str.contains('None')
disord = disord[~ind]
disord
```

```python
disord.to_csv("../../data/disorders/doids.csv")
```
